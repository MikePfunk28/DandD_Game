Actually there is a way, create a test environment using Docker, Inc containers, LocalStack to emulate, and Terraform Labs to turn infrastructure into code.  I would suggest the LocalStack docker extension, with Terraform.

You need to install Docker, I installed the desktop version, which has its own terminal. Then install Localstack, the cli is an executable, same with terraform.

Essentially it works by emulating the AWS services.  Instead of pointing at an AWS service endpoint, you configure the endpoint to point at your localhost, on a port its expecting.

```json
endpoints {
 s3 = "http://localhost:4566"
 # Add other services as needed
 }
```

LocalStack creates a backend and frontend, which has .env file you match to the main.tf file you create. That will be how LocalStack knows, after installing all three. You need LocalStack, which uses docker, and then terraform will aid in automation.

Here is the main.tf file I use for my terraform configuration. You create that in the same directory as terraform.

```json
provider "aws" {
 access_key         = "test"
 secret_key         = "test"
 region           = "us-east-1"
 s3_use_path_style      = true
 skip_credentials_validation = true
 skip_metadata_api_check   = true
 skip_requesting_account_id = true
 endpoints {
 s3 = "http://localhost:4566"
 # Add other services as needed
 }
}

```
After you created the main.tf file, which I would copy exactly an experiment later. Run the tflocal init command in that folder where you created main.tf. This will attempt to initialize it in your local environment.

The last step after running tflocal init and tflocal apply, is to make sure it works. Update your config and credentials file to point to your local env.

hashtag#AWS CLI is wrapped with 'awslocal' which points to your localhost.

You will likely need to set your environment variables in PATH, to where you install the .exe. Follow the instructions on these pages below.

**Note: You need to match your .env with your main.tf. You can then run 'localstack start' and see logs as you create resources.
